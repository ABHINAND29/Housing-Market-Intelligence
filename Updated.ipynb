## MISSING VALUE REPLACEMENT

```python
df.isnull().sum()
```

```python
us=df['Unit_Size_Sqft'].median()
```

```python
df['Unit_Size_Sqft'].fillna(us, inplace=True)
```

```python
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
```

```python
df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median(), inplace=True)
```

```python
df['Amenity_Score'].fillna(df['Amenity_Score'].mean(), inplace=True)
```

```python
df['Buyer_Comments'].fillna("No Comments", inplace=True)
```

```python
df.isnull().sum()
```

## Booking Flag,Price_per_Sqft,Quarter_Number Derivation

```python
df['Booking_Flag'] = df['Possession_Status'].str.strip().str.lower().map({
    'ready to move': 1,
    'launch': 1,
    'under construction': 0
})
```

```python
df['Price_per_Sqft']=df['Ticket_Price_Cr']*1000000/df['Unit_Size_Sqft']
```

```python
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
```

```python
df['Quarter_Number'] = 'Q' + df['Purchase_Quarter'].dt.quarter.astype(str)
```

```python
df.rename(columns={'Developer_Name': 'Builder'}, inplace=True)
```

```python
df = df.drop_duplicates(subset='Property_ID')
```

```python
df['Builder'] = df['Builder'].str.lower()
df['Micro_Market'] = df['Micro_Market'].str.lower()
```
